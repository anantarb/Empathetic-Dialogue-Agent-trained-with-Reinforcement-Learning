```python
import json
preferred_75 = []
preferred_100 = []
preferred_our = []
```

Human Evaluation: Empathy

```python
with open("annotated/ananta_empathy.json") as f:
    ananta_empathy = json.load(f)
with open("annotated/vivi_empathy.json") as f:
    vivi_empathy = json.load(f)
with open("annotated/sophie_empathy.json") as f:
    sophie_empathy = json.load(f)

with open("annotated/ananta_empathy_100.json") as f:
    ananta_empathy_100 = json.load(f)
with open("annotated/vivi_empathy_100.json") as f:
    vivi_empathy_100 = json.load(f)
with open("annotated/sophie_empathy_100.json") as f:
    sophie_empathy_100 = json.load(f)

with open("annotated/ananta_empathy_dialogpt.json") as f:
    ananta_empathy_dialogpt = json.load(f)
with open("annotated/vivi_empathy_dialogpt.json") as f:
    vivi_empathy_dialogpt = json.load(f)
with open("annotated/sophie_empathy_dialogpt.json") as f:
    sophie_empathy_dialogpt = json.load(f)
```

```python
ananta_RL = 0
vivi_RL = 0
sophie_RL = 0

ananta_RL_100 = 0
vivi_RL_100 = 0
sophie_RL_100 = 0

ananta_our = 0
vivi_our = 0
sophie_our = 0

for i in range(len(ananta_empathy['empathy'])):
    if ananta_empathy['empathy'][i]['best'] == 'RL':
        ananta_RL += 1
for i in range(len(vivi_empathy['empathy'])):
    if vivi_empathy['empathy'][i]['best'] == 'RL':
        vivi_RL += 1
for i in range(len(sophie_empathy['empathy'])):
    if sophie_empathy['empathy'][i]['best'] == 'RL':
        sophie_RL += 1

for i in range(len(ananta_empathy_100['empathy'])):
    if ananta_empathy_100['empathy'][i]['best'] == 'RL':
        ananta_RL_100 += 1
for i in range(len(vivi_empathy_100['empathy'])):
    if vivi_empathy_100['empathy'][i]['best'] == 'RL':
        vivi_RL_100 += 1
for i in range(len(sophie_empathy_100['empathy'])):
    if sophie_empathy_100['empathy'][i]['best'] == 'RL':
        sophie_RL_100 += 1

for i in range(len(ananta_empathy_dialogpt['empathy'])):
    if ananta_empathy_dialogpt['empathy'][i]['best'] == 'RL':
        ananta_our += 1
for i in range(len(vivi_empathy_dialogpt['empathy'])):
    if vivi_empathy_dialogpt['empathy'][i]['best'] == 'RL':
        vivi_our += 1
for i in range(len(sophie_empathy_dialogpt['empathy'])):
    if sophie_empathy_dialogpt['empathy'][i]['best'] == 'RL':
        sophie_our += 1


preferred_75.append((ananta_RL + vivi_RL + sophie_RL) / 150 * 100)
preferred_100.append((vivi_RL_100 + sophie_RL_100 + sophie_RL_100)  / 150 * 100)
preferred_our.append((vivi_our + sophie_our + sophie_our)  / 150 * 100)
```

Human Evaluation: Fluency

```python
with open("annotated/ananta_fluency.json") as f:
    ananta_fluency = json.load(f)
with open("annotated/vivi_fluency.json") as f:
    vivi_fluency = json.load(f)
with open("annotated/sophie_fluency.json") as f:
    sophie_fluency = json.load(f)

with open("annotated/ananta_fluency_100.json") as f:
    ananta_fluency_100 = json.load(f)
with open("annotated/vivi_fluency_100.json") as f:
    vivi_fluency_100 = json.load(f)
with open("annotated/sophie_fluency_100.json") as f:
    sophie_fluency_100 = json.load(f)

with open("annotated/ananta_fluency_dialogpt.json") as f:
    ananta_fluency_dialogpt = json.load(f)
with open("annotated/vivi_fluency_dialogpt.json") as f:
    vivi_fluency_dialogpt = json.load(f)
with open("annotated/sophie_fluency_dialogpt.json") as f:
    sophie_fluency_dialogpt = json.load(f)
```

```python
ananta_RL = 0
vivi_RL = 0
sophie_RL = 0

ananta_RL_100 = 0
vivi_RL_100 = 0
sophie_RL_100 = 0

ananta_our = 0
vivi_our = 0
sophie_our = 0

for i in range(len(ananta_fluency['fluency'])):
    if ananta_fluency['fluency'][i]['best'] == 'RL':
        ananta_RL += 1
for i in range(len(vivi_fluency['fluency'])):
    if vivi_fluency['fluency'][i]['best'] == 'RL':
        vivi_RL += 1
for i in range(len(sophie_fluency['fluency'])):
    if sophie_fluency['fluency'][i]['best'] == 'RL':
        sophie_RL += 1


for i in range(len(ananta_fluency_100['fluency'])):
    if ananta_fluency_100['fluency'][i]['best'] == 'RL':
        ananta_RL_100 += 1
for i in range(len(vivi_fluency_100['fluency'])):
    if vivi_fluency_100['fluency'][i]['best'] == 'RL':
        vivi_RL_100 += 1
for i in range(len(sophie_fluency_100['fluency'])):
    if sophie_fluency_100['fluency'][i]['best'] == 'RL':
        sophie_RL_100 += 1

for i in range(len(ananta_fluency_dialogpt['fluency'])):
    if ananta_fluency_dialogpt['fluency'][i]['best'] == 'RL':
        ananta_our += 1
for i in range(len(vivi_fluency_dialogpt['fluency'])):
    if vivi_fluency_dialogpt['fluency'][i]['best'] == 'RL':
        vivi_our += 1
for i in range(len(sophie_fluency_dialogpt['fluency'])):
    if sophie_fluency_dialogpt['fluency'][i]['best'] == 'RL':
        sophie_our += 1


preferred_75.append((ananta_RL + vivi_RL + sophie_RL) / 150 * 100)
preferred_100.append((vivi_RL_100 + sophie_RL_100 + ananta_RL_100)  / 150 * 100)
preferred_our.append((vivi_our + sophie_our + sophie_our)  / 150 * 100)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,5))
plt.bar(['Empathy', 'Relevance', 'Fluency'], preferred_75, color=['blue', 'orange', 'green'], width=0.5)
plt.xlabel('Components')
plt.title('Preference of P75 Model Over Supervised Baseline')
plt.ylabel('Percentage Preferred')
plt.show()
```
