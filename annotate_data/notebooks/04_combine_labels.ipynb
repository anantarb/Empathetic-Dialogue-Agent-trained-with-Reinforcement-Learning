```python
import json

with open('datasets/annotated/ananta.json') as f:
  ananta = json.load(f)

with open('datasets/annotated/vivi.json') as f:
  vivi = json.load(f)

with open('datasets/annotated/sophie.json') as f:
  sophie = json.load(f)
```

```python
same_for_each = 30 # same_for_each represents how many samples were same in each quota
combined = []
combined = combined + ananta['ananta']
combined = combined + vivi['vivi'][same_for_each:]
combined = combined + sophie['sophie'][same_for_each:]
print(len(combined))
```

```python
with open('datasets/combined_labels.json', 'w') as outfile:
    json.dump(combined, outfile)
```
